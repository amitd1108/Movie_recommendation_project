```python
import ast 
merged_df.genres=merged_df.genres.apply( lambda x: ast.literal_eval(x))
```

```python
merged_df.genres.value_counts()
```

```python
def access_genre(x):
    gen_list=[]
    for i in x:
        y=i['name']
        if y=='Science Fiction':
            y='Sci-Fi'
        gen_list.append(y)
    if gen_list==[]:
        return(np.nan)
    else :
        return(' '.join(gen_list))
        
        
```

```python
merged_df.genres=merged_df.genres.apply(access_genre)
```

```python
merged_df.head()
```

```python
merged_df.cast=merged_df.cast.apply( lambda x: ast.literal_eval(x))
```

```python
merged_df.cast.iloc[1]
```

```python
# taking top 3 casts 
def access_cast(x):
    cast_list=[]
    for i in x:
        cast_list.append(i['name'])
    return(cast_list[:3])
        
```

```python
merged_df.crew.iloc[1]
```

```python
merged_df.crew=merged_df.crew.apply( lambda x: ast.literal_eval(x))
```

```python
# taking top Director 
def access_director(x):
    director_list=[]
    for i in x:
        if i['job']=='Director':
            director_list.append(i['name'])
    return(director_list[:1])
```

```python
merged_df['director']=merged_df.crew.apply(access_director)
```

```python
merged_df.cast=merged_df.cast.apply(access_cast)
```

```python
merged_df=merged_df[(merged_df.director.apply(len)>0) & (merged_df.cast.apply(len)==3)]
```

```python
merged_df.cast=merged_df.cast.apply(lambda x:' '.join(x))
```

```python
merged_df.director=merged_df.director.apply(lambda x: x[0])
```

```python
merged_df['comb']= merged_df.cast + ' ' +  merged_df.director +  ' ' + merged_df.genres
```

```python
new_df=merged_df.drop(['id','imdb_id','overview','spoken_languages','vote_average','vote_count'
                       ,'cast','crew'],axis=1)
new_df.head()
```

```python
print(new_df.shape)
```

```python
print(new_df.comb.dtype)
```

```python
new_df.comb=new_df.comb.astype('str').apply( lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
comb_vec=vectorizer.fit_transform(new_df.comb).toarray()
```

```python
comb_vec.shape
```

```python
comb_vec
```

```python
print(vectorizer.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(comb_vec)
```

```python
similarity.shape
```

```python
def recommend_movie(movie):
    if movie not in new_df['title'].unique():
        return('Sorry! The movie you requested is not in our database. Please check the spelling or try with some other movies')
    else:
        movie_index=new_df[new_df.title==movie].index[0]
        cos_similarity=similarity[movie_index]
    
        list1=sorted(enumerate(cos_similarity),reverse=True,key=(lambda x : x[1]))
        for i in range (1,6):
            print(new_df.title.iloc[list1[i][0]] +' ,similarity='+ str(list1[i][1])) 
```

```python
recommend_movie('The Revenant')
```
